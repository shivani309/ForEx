```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import *
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from sklearn.metrics import (
    mean_absolute_error, 
    mean_squared_error, 
    mean_absolute_percentage_error, 
    r2_score
)
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
X = pd.read_csv(r'C:/Users/uzmap/Documents/GitHub/ForEx/USDINR/X_USDINR.csv')
X.drop('Unnamed: 0', axis=1, inplace=True)
X.head()
```

```python
y = pd.read_csv(r'C:/Users/uzmap/Documents/GitHub/ForEx/USDINR/Y_USDINR.csv')
y.drop('Unnamed: 0', axis=1, inplace=True)
y.head()
```

```python
def gradient_boosting_model(X,y):
    
    gb = GradientBoostingRegressor()
    gb_metrics = {}

    # Prepare data for training and testing
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
        
    # Train the model
    gb.fit(X_train, y_train)
    
    # Predict
    y_test_pred = gb.predict(X_test)
    y_train_pred = gb.predict(X_train)
    
    # Calculate metrics
    testmae = mean_absolute_error(y_test, y_test_pred)
    testmse = mean_squared_error(y_test, y_test_pred)
    testmape = mean_absolute_percentage_error(y_test, y_test_pred)
    testr2 = r2_score(y_test, y_test_pred)

    trainmae = mean_absolute_error(y_train, y_train_pred)
    trainmse = mean_squared_error(y_test, y_test_pred)
    trainmape = mean_absolute_percentage_error(y_train, y_train_pred)
    trainr2 = r2_score(y_train, y_train_pred)

    # Plot actual vs predicted trends
    plt.figure(figsize=(10, 6))
    plt.plot(np.arange(len(y_test)), y_test, label="Actual Price", color="blue")
    plt.plot(np.arange(len(y_test_pred)), y_test_pred, label="Predicted Price", color="red", linestyle="dashed")
    plt.title("Actual vs Predicted Price Trends")
    plt.xlabel("Index")
    plt.ylabel("Price")
    plt.legend()
    plt.show()
    
    # After calculating the metrics:
    gb_metrics = {
        "Metric": [
            "Mean Absolute Error (MAE)", 
            "Mean Squared Error (MSE)",
            "Mean Absolute Percentage Error (MAPE)", 
            "R-squared (R)", 
            "Accuracy (1 - MAPE)"
        ],
        "Training": [
            trainmae,
            trainmse,
            trainmape * 100, 
            trainr2, 
            (1 - trainmape) * 100
        ],
        "Testing": [
            testmae, 
            testmse,
            testmape * 100, 
            testr2, 
            (1 - testmape) * 100
        ]
    }

    # Convert to DataFrame
    gb_metrics_df = pd.DataFrame(gb_metrics)
    return gb_metrics_df


gb_metrics_df = gradient_boosting_model(X,y)
```

with CrossValidation

```python
def gradient_boosting_model_cv(X, y):
    gb = GradientBoostingRegressor()
    
    # Define the cross-validation strategy
    cv = KFold(n_splits=5, shuffle=True, random_state=42)  # 5-fold CV

    # Perform cross-validation for MAE and MSE
    mae_scores = -cross_val_score(gb, X, y, scoring='neg_mean_absolute_error', cv=cv)
    mse_scores = -cross_val_score(gb, X, y, scoring='neg_mean_squared_error', cv=cv)
    mape_scores = []
    r2_scores = []

    for train_idx, test_idx in cv.split(X):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        
        # Fit the model
        gb.fit(X_train, y_train)
        
        # Predict
        y_pred = gb.predict(X_test)
        
        # Calculate additional metrics
        mape_scores.append(mean_absolute_percentage_error(y_test, y_pred))
        r2_scores.append(r2_score(y_test, y_pred))

    # Average the metrics across folds
    mean_mae = np.mean(mae_scores)
    mean_mse = np.mean(mse_scores)
    mean_mape = np.mean(mape_scores)
    mean_r2 = np.mean(r2_scores)

    # After calculating the metrics:
    gb_metrics_cv = {
        "Metric": [
            "Mean Absolute Error (MAE)", 
            "Mean Squared Error (MSE)",
            "Mean Absolute Percentage Error (MAPE)", 
            "R-squared (R)", 
            "Accuracy (1 - MAPE)"
        ],
        "Cross-Validation Scores": [
            mean_mae, 
            mean_mse, 
            mean_mape * 100, 
            mean_r2, 
            (1 - mean_mape) * 100
        ]
    }

    # Convert to DataFrame
    gb_metrics_cv_df = pd.DataFrame(gb_metrics_cv)
    return gb_metrics_cv_df


# Call the function with your data
gb_metrics_cross_validation = gradient_boosting_model_cv(X, y)
```
